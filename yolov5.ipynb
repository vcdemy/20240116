```python
import cv2
import torch
from PIL import Image
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
im = Image.open('beattles.jpg')  # PIL image

# Inference
results = model(im, augment=True)  # batch of images
```

```python
results.show()
```
