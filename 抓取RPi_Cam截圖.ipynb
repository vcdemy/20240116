# 抓取RPi Webcam的截圖

```python
import requests
```

```python
from PIL import Image
```

```python
url = "http://192.168.50.199:5000/snapshot"
```

```python
r = requests.get(url)
with open('captured.jpg', 'wb') as f:
    f.write(r.content)
Image.open('captured.jpg')
```
